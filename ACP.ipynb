```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.plotly as py
import plotly.tools as tls
from plotly.graph_objs import *

import pandas as pd
# py.iplot_mpl(fig, filename='testipy')
```

```python
exe_47 = pd.read_csv('../acp/array_47/executions.csv')
exe_47['StartTime'] = exe_47.apply(lambda r: pd.datetime.strptime(r['StartTime'], '%Y-%m-%dT%H:%M:%S'), axis=1)
exe_47['EndTime'] = exe_47.apply(lambda r: pd.datetime.strptime(r['EndTime'], '%Y-%m-%dT%H:%M:%S'), axis=1)
```

```python
exe_47.head()
```

```python
sns.set_style("darkgrid")
fig, ax = plt.subplots()

sb_tp = ax.plot_date(exe_47.query('ArrayType == "TP_ARRAY"').StartTime,
                     exe_47.query('ArrayType == "TP_ARRAY"').Band, 'o', mfc='w', mec='r', mew=1,
                     label='TP-Array')
sb_7 = ax.plot_date(exe_47.query('ArrayType == "SEVEN_M"').StartTime,
                     exe_47.query('ArrayType == "SEVEN_M"').Band, 'o', mfc='w', mec='k', mew=1,
                     label='Seven-M')
sb_12 = ax.plot_date(exe_47.query('ArrayType == "TWELVE_M"').StartTime,
                     exe_47.query('ArrayType == "TWELVE_M"').Band, 'o', mfc='w', mec='b', mew=1,
                     label='Twelve-M')

ax.set_ylim(0,10)
ax.set_xlabel(r'Date')
ax.set_ylabel(r'Band')


py_fig = tls.mpl_to_plotly(fig, resize=False)
py_fig['layout'].pop('annotations', None)

py_fig['layout'].update(
    showlegend=True,
    legend=Legend(
        x=1.05,
        y=1
    )
)

py_fig['layout']['height']  = 500
py_fig['layout']['width'] = 800
py_fig['layout']['autosize'] = False

py.iplot(py_fig, filename='executions_47')
```
